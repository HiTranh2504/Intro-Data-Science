```python
#Câu 1 Đọc dữ liệu từ tệp iris.data trong Google Colab
from google.colab import files
uploaded = files.upload()
data = []
with open('/content/iris.data.txt', 'r') as file:
  lines = file.readlines()
  for line in lines:
    data.append(line.strip().split(','))
```

```python
#Câu 2 Mã hóa cột class thành các con số tương ứng
class_mapping = {'Iris-setosa': 1, 'Iris-versicolor': 2, 'Iris-virginica': 3}

for i in range(len(data)):
    label = data[i][-1]
    if label in class_mapping:
        data[i][-1] = class_mapping[label]

#In dữ liệu
data
```

```python

#Câu 3 Tính giá trị lớn nhất, nhỏ nhất và trung bình
def calculate_max_min_avg(column):
  max_value = column[0]
  min_value = column[0]
  sum_value = 0
  for value in column:
    max_value = max(max_value, value)
    min_value = min(min_value, value)
    sum_value += value
  avg_value = sum_value / len(column)
  return max_value, min_value, avg_value
```

```python
#Câu 4: Áp dụng hàm tính giá trị lớn nhất, nhỏ nhất và trung bình cho từng cột
column_1 = [float(row[0]) if len(row) > 0 and row[0] != '' else 0.0 for row in data]
column_2 = [float(row[1]) if len(row) > 1 and row[1] != '' else 0.0 for row in data]
column_3 = [float(row[2]) if len(row) > 2 and row[2] != '' else 0.0 for row in data]
column_4 = [float(row[3]) if len(row) > 3 and row[3] != '' else 0.0 for row in data]
max_1, min_1, avg_1 = calculate_max_min_avg(column_1)
max_2, min_2, avg_2 = calculate_max_min_avg(column_2)
max_3, min_3, avg_3 = calculate_max_min_avg(column_3)
max_4, min_4, avg_4 = calculate_max_min_avg(column_4)

# Hiển thị kết quả ra màn hình
print("Cột 1:")
print("Giá trị lớn nhất:", max_1)
print("Giá trị nhỏ nhất:", min_1)
print("Giá trị trung bình:", avg_1)

print("Cột 2:")
print("Giá trị lớn nhất:", max_2)
print("Giá trị nhỏ nhất:", min_2)
print("Giá trị trung bình:", avg_2)

print("Cột 3:")
print("Giá trị lớn nhất:", max_3)
print("Giá trị nhỏ nhất:", min_3)
print("Giá trị trung bình:", avg_3)

print("Cột 4:")
print("Giá trị lớn nhất:", max_4)
print("Giá trị nhỏ nhất:", min_4)
print("Giá trị trung bình:", avg_4)
```

```python
#Câu 5 Viết hàm tính Độ lệch chuẩn (standard deviation) theo công thức sau:
import numpy as np
def tinh_do_lech_chuan(data):
    # Tính giá trị trung bình
    mean = np.mean(data)
    # Tính độ lệch từng phần tử so với giá trị trung bình
    deviations = data - mean
    # Tính bình phương của độ lệch
    squared_deviations = deviations ** 2
    # Tính độ lệch chuẩn bằng cách lấy căn bậc hai của giá trị trung bình của bình phương độ lệch
    standard_deviation = np.sqrt(np.mean(squared_deviations))
    return standard_deviation
```

```python
#Câu 6 Áp dụng 3 hàm vừa viết ở 5 vào từng cột 1,2,3,4 và hiển thị kết quả ra màn hình.
# Apply the function to each column
std_dev_1 = tinh_do_lech_chuan(column_1)
std_dev_2 = tinh_do_lech_chuan(column_2)
std_dev_3 = tinh_do_lech_chuan(column_3)
std_dev_4 = tinh_do_lech_chuan(column_4)
# Print the results
print("Cột 1:")
print("Độ lệch chuẩn:", std_dev_1)
print("Cột 2:")
print("Độ lệch chuẩn:", std_dev_2)
print("Cột 3:")
print("Độ lệch chuẩn:", std_dev_3)
print("Cột 4:")
print("Độ lệch chuẩn:", std_dev_4)
```
