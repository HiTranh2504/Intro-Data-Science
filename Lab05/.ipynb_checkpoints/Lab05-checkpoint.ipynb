### Bài 1

```python
class Animal:
    def __init__(self, picture, food, hunger, boundaries, location):
        self.picture = picture  
        self.food = food
        self.hunger = hunger
        self.boundaries = boundaries    
        self.location = location
        
    def make_noise(self):
        print("The animal makes noise.")
        
    def eat_food(self):
        print("The animal is eating.")
        
    def is_hungry(self):
        print("The animal is hungry.")
        
    def has_boundaries(self):
        print("The animal has boundaries.")
        
    def get_location(self):
        print("The animal is located at " + self.location)
        

class Lion(Animal):
    def __init__(self, picture, food, hunger, boundaries, location):
        super().__init__(picture, food, hunger, boundaries, location)
        
    def make_noise(self):
        super().make_noise()
        print("Roar")
        
    def eat_food(self):
        super().eat_food()
        print("meat")
        

class Cat(Animal):
    def __init__(self, picture, food, hunger, boundaries, location):
        super().__init__(picture, food, hunger, boundaries, location)
        
    def make_noise(self):
        super().make_noise()
        print("Meow Meow")
        
    def eat_food(self):
        super().eat_food()
        print("fish")
        

class Tiger(Animal):
    def __init__(self, picture, food, hunger, boundaries, location):
        super().__init__(picture, food, hunger, boundaries, location)
        
    def make_noise(self):
        super().make_noise()
        print("Roar roar")
        
    def eat_food(self):
        super().eat_food()
        print("Thit heo")
        

class Hippo(Animal):
    def __init__(self, picture, food, hunger, boundaries, location):
        super().__init__(picture, food, hunger, boundaries, location)
        
    def make_noise(self):
        super().make_noise()
        print("Roar roar")
        
    def eat_food(self):
        super().eat_food()
        print("Thit heo")
        

class Canine(Animal):
    def __init__(self, picture, food, hunger, boundaries, location):
        super().__init__(picture, food, hunger, boundaries, location)
        
    def roam(self):
        print("The animal is roaming.")
        

class Dog(Canine):
    def __init__(self, picture, food, hunger, boundaries, location):
        super().__init__(picture, food, hunger, boundaries, location)
        
    def make_noise(self):
        super().make_noise()
        print("Gau Gau")
        
    def eat_food(self):
        super().eat_food()
        print("xuong")
        
    def chase_cat(self):
        print("The dog is chasing a cat.")
        

class Wolf(Canine):
    def __init__(self, picture, food, hunger, boundaries, location):
        super().__init__(picture, food, hunger, boundaries, location)
        
    def make_noise(self):
        super().make_noise()
        print("Gau Gau")
        
    def eat_food(self):
        super().eat_food()
        print("xuong")
        

def main():
    lion = Lion("lion.jpg", "meat", "hungry", "forest", "Africa")
    lion.make_noise()
    lion.eat_food()
    print(lion.boundaries)
    lion.get_location()
    
    cat = Cat("cat.jpg", "fish", "hungry", "house", "Vietnam")
    cat.make_noise()
    cat.eat_food()
    print(cat.boundaries)
    cat.get_location()
    
    tiger = Tiger("tiger.jpg", "meat", "hungry", "forest", "India")
    tiger.make_noise()
    tiger.eat_food()
    print(tiger.boundaries)
    tiger.get_location()
    
    hippo = Hippo("hippo.jpg", "grass", "hungry", "river", "Africa")
    hippo.make_noise()
    hippo.eat_food()
    print(hippo.boundaries)
    hippo.get_location()
    
    dog = Dog("dog.jpg", "xuong", "hungry", "house", "Vietnam")
    dog.make_noise()
    dog.eat_food()
    print(dog.boundaries)
    dog.get_location()
    dog.chase_cat()
    
    wolf = Wolf("wolf.jpg", "xuong", "hungry", "forest", "Vietnam")
    wolf.make_noise()
    wolf.eat_food()
    print(wolf.boundaries)
    
main()
```


### Bài 3

```python
class Expression:
    def __init__(self, left, right):
        self.left = left
        self.right = right
    def evaluate(self):
        pass
    def toString(self):
        pass
    
class Number(Expression):
    def __init__(self, value):
        self.value = value
    def evaluate(self):
        return self.value
    def toString(self):
        return str(self.value)
    
class Square(Expression):
    def __init__(self, value):
        self.value = value
    def evaluate(self):
        return self.value * self.value
    def toString(self):
        return str(self.value) + " * " + str(self.value)
    
class BinaryXpression(Expression):
    def __init__(self, left, right):
        self.left = left
        self.right = right
    def evaluate(self):
        pass
    def toString(self):
        pass
    
class Addition(BinaryXpression):
    def __init__(self, left, right):
        self.left = left
        self.right = right
    def evaluate(self):
        return self.left.evaluate() + self.right.evaluate()
    def toString(self):
        return self.left.toString() + " + " + self.right.toString()
    
def main():
    number = Number(5)
    print(number.toString())
    print(number.evaluate())
    
    square = Square(5)
    print(square.toString())
    print(square.evaluate())
    
    addition = Addition(Number(5), Number(6))
    print(addition.toString())
    print(addition.evaluate())
    
main()
```
