```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#### DATA VISUALIZATION

```python
#1.9. Visualize the top 10 popular names in 2014
df2014 = df[df['Year'] == 2014]
top10namein2014 = df2014['Name'].value_counts().head(10)
top10namein2014

plt.figure(figsize=(10, 6))
top10namein2014.plot(kind='bar', color='pink')
plt.title('10 tên phổ biến nhất năm 2014')
plt.xlabel('Tên')
plt.ylabel('Số lần xuất hiện')
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.show()
```

```python
#1.10. Visualize the ratio of men and women over the years 2004-2014
df04_14 = df[df['Year'].between(2004, 2014)]

# Tính tỉ lệ nam nữ cho mỗi năm
gender_ratio = df04_14.groupby('Year')['Gender'].value_counts(normalize=True).unstack()

# Trực quan hóa biểu đồ đường
plt.figure(figsize=(10, 6))
gender_ratio.plot(kind='line', marker='o')
plt.title('Tỉ lệ nam nữ từ năm 2004 đến 2014')
plt.xlabel('Năm')
plt.ylabel('Tỉ lệ')
plt.xticks(range(2004, 2015), rotation=45)
plt.legend(title='Giới tính', loc='upper right')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#2.12. What is the tip to total bill?
total_bill = df['total_bill'].value_counts()
total_bill
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x=df['total_bill'], data=df)
plt.title("Rate total bill")
plt.show()
```

```python
#2.13. Between men and women, who tips more money, who tips more often?

# Tính tổng số lần tiếp theo giới tính
total_visits_by_gender = df['sex'].value_counts()

# Tính trung bình tiền tip theo giới tính
average_tip_by_gender = df.groupby('sex')['tip'].mean()

# In kết quả
print("Tổng số lần tiếp theo giới tính:")
print(total_visits_by_gender)

print("\nTrung bình tiền tip theo giới tính:")
print(average_tip_by_gender)
```

```python
#2.14. Between smokers and non-smokers, who tips more, who tips more often?

# Tính tổng số lần tiếp theo giới tính
total_visits_by_gender = df['smoker'].value_counts()

# Tính trung bình tiền tip theo giới tính
average_tip_by_gender = df.groupby('smoker')['tip'].mean()

# In kết quả
print("Tổng số lần tiếp theo hút thuốc:")
print(total_visits_by_gender)

print("\nTrung bình tiền tip theo hút thuốc:")
print(average_tip_by_gender)
```

```python
#2.15. Statistics of the total amount of tips received by day.
total_tip_by_day = df.groupby('day')['tip'].sum()
total_tip_by_day
```

### DATA VISUALIZATION

```python
#2.16. Visualize total tips by day
total_tip_by_day = df.groupby('day')['tip'].sum().reset_index()
sns.barplot(x=total_tip_by_day['day'], y=total_tip_by_day['tip'])
plt.xlabel('Day')
plt.ylabel('Total Tips')
plt.title('Total Tips by Day')
plt.show()
```

```python
#2.17. Visualize total tip by meal (time)
total_tip_by_meals = df.groupby('time')['tip'].sum().reset_index()
sns.barplot(x=total_tip_by_meals['time'], y=total_tip_by_meals['tip'])
plt.xlabel('Time')
plt.ylabel('Total Tips')
plt.title('Total Tips by Time')
plt.show()
```

```python
#2.18. Plot histogram of the total invoice
plt.hist(df['total_bill'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.title('Histogram of Total Invoice')
plt.show()
```

```python
#2.19. Plot the boxplot of the total invoice
sns.boxplot(x=df['total_bill'])
plt.xlabel('Total Bill')
plt.title('Boxplot of Total Invoice')
plt.show()
```

```python
#2.20. Visualize the relationship between total bill and tip
sns.scatterplot(data=df, x='total_bill', y='tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Relationship Between Total Bill And Tip')
plt.show()
```

```python
#2.21. Compare how much customers have spent between days of the week
total_bill_by_day = df.groupby('day')['total_bill'].sum().reset_index()
sns.barplot(x=total_bill_by_day['day'], y=total_bill_by_day['total_bill'])
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.title('Total Bill by Day')
plt.show()
```
