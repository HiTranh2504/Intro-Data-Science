## Bài 2
Viết chương trình sử dụng cấu trúc dữ liệu Danh sách liên kết đơn – SLL quản lý Tàu

6. depart_place (string): Nơi khởi hành

Viết chương trình sử dụng cấu trúc dữ liệu Danh sách liên kết đơn – SLL quản lý Tàu 
(Traines) theo thông tin như 
Train:
* 1. tcode (string): mã của tàu (mỗi tàu có một mã duy nhất).
* 2. train_name (string): tên của tàu.
* 3. seat (integer): số ghế trên tàu (seat > 0).
* 4. booked (integer): số ghế đã đặt (booked >= 0 and booked ≤ seat). 
* 5. depart_time (double): Thời gian khởi hành (depart_time >= 0).
* 6. depart_place (string): Nơi khởi hành.
* Thực hiện các yêu cầu như dưới đây, chú ý phải xử lý Exception cho toàn chương trình:
* 2.1. Đọc dữ liệu từ file input.txt có nội dung như sau:
- B03 | Sug | 12 | 3 | 11 | PA 
- B01 | Mil | 10 | 5 | 5.7 | PC
- B02 | App | 5 | 2 | 4 | PB
- B05 | Roo | 7 | 6 | 15 | PE
- B07 | Bee | 11 | 3 | 12 | PF
- B04 | Boo | 9 | 5 | 5 | PD
* Dòng đầu tiên có ý nghĩa: tcode = B03, train_name = Sug, seat = 12, booked = 3, depart_time = 11 và depart_place = PA
* 2.2. Nhập và thêm dữ liệu vào cuối danh sách
* 2.3. Hiển thị dữ liệu
* 2.4. Lưu dữ liệu vào 1 file
* 2.5. Tìm kiếm dữ liệu theo tcode
* 2.6. Xóa dữ liệu theo tcode
* 2.7. Sắp xếp dữ liệu theo tcode
* 2.8. Thêm 1 dữ liệu sau vị trí k
* 2.9. Sửa dữ liệu theo tcode
* 2.10. Xóa 1 node trước node có tcode = xCode 

```python
class Train:
    def __init__(self, tcode, train_name, seat, booked, depart_time, depart_place):
        if not isinstance(tcode, str):
            raise ValueError("Mã tàu phải là một chuỗi.")
        if not isinstance(train_name, str):
            raise ValueError("Tên tàu phải là một chuỗi.")
        
        self.seat = seat
        if not isinstance(self.seat, int) or self.seat <= 0:
            raise ValueError("số ghế trên tàu > 0")
        
        self.booked = booked
        if not isinstance(self.booked, int) or self.booked < 0 or self.booked > self.seat:
            raise ValueError("booked >= 0 and booked ≤ seat")
        
        if not isinstance(depart_time, (int, float)) or depart_time < 0:
            raise ValueError("Thời gian khởi hành >= 0.")
        if not isinstance(depart_place, str):
            raise ValueError("Nơi khởi hành phải là một chuỗi.")

        self.tcode = tcode
        self.train_name = train_name
        self.depart_time = depart_time
        self.depart_place = depart_place
        self.next = None
        self.prev = None


class TrainDLL:
    def __init__(self):
        self.head = None

    def append(self, train):
        existing_train = self.search_by_tcode(train.tcode)
        if existing_train:
            print(f"Tàu với mã {train.tcode} đã tồn tại.")
            return False

        if not self.head:
            self.head = train
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = train
            train.prev = current
        return True

    def display(self):
        current = self.head
        while current:
            print(f"{current.tcode} | {current.train_name} | {current.seat} | {current.booked} | {current.depart_time} | {current.depart_place}")
            current = current.next

    def save_to_file(self, filename):
        with open(filename, 'w') as file:
            current = self.head
            while current:
                file.write(f"{current.tcode} | {current.train_name} | {current.seat} | {current.booked} | {current.depart_time} | {current.depart_place}\n")
                current = current.next

    def search_by_tcode(self, tcode):
        current = self.head
        while current:
            if current.tcode == tcode:
                return current
            current = current.next
        return None

    def delete_by_tcode(self, tcode):
        current = self.head
        while current:
            if current.tcode == tcode:
                if current.prev:
                    current.prev.next = current.next
                else:
                    self.head = current.next

                if current.next:
                    current.next.prev = current.prev

                del current
                break
            current = current.next

    def sort_by_tcode(self):
        trains = []
        current = self.head
        while current:
            trains.append(current)
            current = current.next

        sorted_trains = sorted(trains, key=lambda x: x.tcode)
        
        self.head = sorted_trains[0]
        current = self.head
        for train in sorted_trains[1:]:
            current.next = train
            train.prev = current
            current = current.next
        current.next = None
        
    def insert_after_k(self, k, train):
        current = self.head
        while current:
            if current.tcode == k:
                train.next = current.next
                train.prev = current
                if current.next:
                    current.next.prev = train
                current.next = train
                break
            current = current.next

    def update_by_tcode(self, tcode, new_train):
        current = self.head
        while current:
            if current.tcode == tcode:
                current.tcode = new_train.tcode
                current.train_name = new_train.train_name
                current.seat = new_train.seat
                current.booked = new_train.booked
                current.depart_time = new_train.depart_time
                current.depart_place = new_train.depart_place
                break
            current = current.next

    def delete_before_xcode(self, xcode):
        current = self.head
        while current.next:
            if current.next.tcode == xcode:
                if current.prev:
                    current.prev.next = current.next
                    current.next.prev = current.prev
                else:
                    self.head = current.next
                    current.next.prev = None
                del current
                break
            current = current.next

def read_data_from_file(filename):
    train_dll = TrainDLL()
    try:
        with open(filename, 'r') as file:
            lines = file.readlines()
            for line in lines:
                data = line.strip().split(" | ")
                train = Train(data[0], data[1], int(data[2]), int(data[3]), float(data[4]), data[5])
                train_dll.append(train)
    except Exception as e:
        print(f"Error reading file: {e}")
    return train_dll


def main():
    try:
        train_dll = read_data_from_file("input.txt")

        while True:
            print("1. Hiển thị Danh sách Tàu")
            print("2. Thêm Tàu mới")
            print("3. Lưu Danh sách Tàu vào Tệp")
            print("4. Tìm kiếm Tàu theo mã tàu")
            print("5. Xóa Tàu theo mã tàu")
            print("6. Sắp xếp Danh sách Tàu theo mã tàu")
            print("7. Thêm Tàu mới sau mã tàu")
            print("8. Cập nhật thông tin Tàu theo mã tàu")
            print("9. Xóa Node trước mã tàu")
            print("0. Thoát")

            choice = int(input("Nhập lựa chọn: "))
            
            if choice == 0:
                break

            elif choice == 1:
                print("\nTrain list:")
                train_dll.display()

            elif choice == 2: 
                tcode = input("Nhập tcode: ")
                train_name = input("Nhập train name: ")
                seat = int(input("Nhập seat: "))
                booked = int(input("Nhập booked: "))
                depart_time = float(input("Nhập depart time: "))
                depart_place = input("Nhập depart place: ")
                new_train = Train(tcode, train_name, seat, booked, depart_time, depart_place)
                tmp = train_dll.append(new_train)
                if tmp:
                  print("Thêm thành công!")

            elif choice == 3:
                train_dll.save_to_file("output.txt")
                print("Train List saved to Output.txt")

            elif choice == 4:
                tcode_to_search = input("Nhập mã muốn tìm: ")
                searched_train = train_dll.search_by_tcode(tcode_to_search)
                if searched_train:
                    print(f"{searched_train.tcode} | {searched_train.train_name} | {searched_train.seat} | {searched_train.booked} | {searched_train.depart_time} | {searched_train.depart_place}")
                else:
                    print(f"\nNot found {tcode_to_search}.")
            
            elif choice == 5:
                tcode_to_delete = input("Nhập mã muốn xóa: ")
                train_dll.delete_by_tcode(tcode_to_delete)
                print(f"Đã xóa {tcode_to_delete}")

            elif choice == 6:
                train_dll.sort_by_tcode()
                train_dll.display()

            elif choice == 7:
                tcode_to_insert_after = input("Nhập tcode chèn sau: ")
                new_train_after_k =  Train(input("Nhập mã tàu mới: "),
                                      input("Nhập tên tàu mới: "),
                                      int(input("Nhập số chỗ mới: ")),
                                      int(input("Nhập số chỗ đã đặt mới: ")),
                                      float(input("Nhập giờ khởi hành mới: ")),
                                      input("Nhập nơi khởi hành mới: "))
                train_dll.insert_after_k(tcode_to_insert_after, new_train_after_k)
                print(f"Tàu mới đã được chèn sau mã tàu {tcode_to_insert_after}.")

            elif choice == 8:
                tcode_to_update = input("Enter tcode to update: ")
                updated_train = Train(tcode_to_update,
                                      input("Nhập tên tàu mới: "),
                                      int(input("Nhập số chỗ mới: ")),
                                      int(input("Nhập số chỗ đã đặt mới: ")),
                                      float(input("Nhập giờ khởi hành mới: ")),
                                      input("Nhập nơi khởi hành mới: "))
                train_dll.update_by_tcode(tcode_to_update, updated_train)
                print("\nCập nhật thành công")

            elif choice == 9:
                xcode_to_delete_before = input("Nhập mã tàu để xóa Node trước nó: ")
                train_dll.delete_before_xcode(xcode_to_delete_before)
                print(f"Xóa thành công node trước node có mã tàu {xcode_to_delete_before}")

            else:
                print("Lựa chọn không hợp lệ. Vui lòng nhập lại (0-9)")

    except Exception as e:
        print(f"Error {e}")
        

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 3.1 Đọc dữ liệu và hiển thị 20 dòng đầu tiên của dữ liệu
data = pd.read_csv('/car (1).csv')
print("20 dòng đầu tiên của dữ liệu:")
print(data.head(20))

# Kiểu dữ liệu của mỗi cột
print("\nKiểu dữ liệu của mỗi cột:")
print(data.dtypes)

# 3.2 Dữ liệu còn trống (null) trong mỗi cột
print("\nDữ liệu còn trống trong mỗi cột:")
print(data.isnull().sum())

# Print column names to identify the correct ones
print("\nColumn names:")
print(data.columns)

# 3.3 Thông tin về số lượng xe của mỗi loại
car_count_by_type = data['car_type'].value_counts()
print("\nSố lượng xe của mỗi loại:")
print(car_count_by_type)

# 3.4 Số lượng xe mới/cũ theo từng loại xe
new_old_count = data.groupby(['car_type', 'new_old']).size().unstack()
print("\nSố lượng xe mới/cũ theo từng loại xe:")
print(new_old_count)

# 3.5 Vẽ biểu đồ số lượng xe mới/cũ qua từng năm
sns.set()
plt.figure(figsize=(12, 6))
sns.countplot(x='car_year', hue='new_old', data=data)
plt.title('Số lượng xe mới/cũ qua từng năm')
plt.show()

# 3.6 Tạo bảng dữ liệu mới chỉ có 3 vùng (TP HCM, Hà Nội và Đà Nẵng)
selected_regions = ['TP HCM', 'Ha Noi', 'Da Nang']
new_data = data[data['area'].isin(selected_regions)]
new_data.to_csv('new_data.csv', index=False)

# 3.7 So sánh các đại lượng thống kê giá trị trung bình, trung vị, độ lệch chuẩn của giá xe
stats_comparison = new_data.groupby('area')['price'].agg(['mean', 'median', 'std'])
print("\nSo sánh giá trung bình, trung vị và độ lệch chuẩn của giá xe:")
print(stats_comparison)

# 3.8 Vẽ đồ thị phân phối của các loại màu sắc
plt.figure(figsize=(10, 6))
sns.countplot(x='in_color', data=data)
plt.title('Phân phối các loại màu sắc')
plt.show()

# 3.9 Vẽ mối tương quan giữa giá xe và số km đã đi đối với xe cũ
old_cars = data[data['new_old'] == 'Old']
sns.scatterplot(x='price', y='km', data=old_cars)
plt.title('Mối tương quan giữa giá xe và số km đã đi cho xe cũ')
plt.show()

# 3.10 Khảo sát mối quan hệ giữa giá của xe SUV và Sedan theo từng năm
suv_sedan_prices = data[data['car_type'].isin(['SUV', 'Sedan'])]
plt.figure(figsize=(12, 6))
sns.lineplot(x='car_year', y='price', hue='car_type', data=suv_sedan_prices)
plt.title('Mối quan hệ giữa giá của xe SUV và Sedan theo từng năm')
plt.show()
```
