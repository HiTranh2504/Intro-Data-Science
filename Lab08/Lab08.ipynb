```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#1. Đọc dữ liệu và hiển thị ra 20 mẫu đầu tiên
df = pd.read_csv('Lab8-pima.csv')
df.head(20)
```

```python
#2. Cho biết Kích thước dữ liệu
df.shape
```

```python
#3. Cho biết Kiểu dữ liệu
df.dtypes
```

```python
#4. Cho biết số lượng mỗi lớp có trong dữ liệu
df.info()
```

```python
#5. Thực hiện các thống kê cơ bản về các trường trong bộ dữ liệu
df.describe()
```

```python
#6. Trực quan hóa dữ liệu với các biểu đồ grid line, box, histogram, và scatter matrix

#6.1 Biểu đồ grid line
ax = df.plot(kind='line')
ax.set_title('Biểu đồ đường')
ax.set_xlabel('Nhóm')
ax.set_ylabel('Giá trị')
ax.grid(True)  # Thêm grid line
plt.show()
```

```python
#6.2 Biểu đồ box
ax = df.boxplot()
ax.set_title('Biểu đồ hộp của các cột')
ax.set_ylabel('Giá trị')
plt.show()
```

```python
#6.3 Biểu đồ histogram
df.hist(bins=20, figsize=(10, 8), alpha=0.7)
plt.suptitle('Biểu đồ histogram')
plt.show()

```

```python
#6.4 Biểu đồ scatter matrix
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix

# Tạo biểu đồ scatter matrix
scatter_matrix(df, figsize=(10, 10), alpha=0.7, diagonal='kde')
plt.suptitle('Biểu đồ scatter matrix')
plt.show()
```

```python
#7. Tính mối tương quan Pearson cho từng cặp thuộc tính và trực quan hóa kết quả
corr = df.corr(method='pearson')
plt.figure(figsize=(20, 20))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=2)
plt.show()
```
