```python
import scrapy
from bs4 import BeautifulSoup
import json
from scrapy import FormRequest

def soup_to_scheduler(soup):
    schedule = {}
    for row in soup.find_all('tr')[1:]:  # Bỏ qua hàng đầu tiên chứa tiêu đề
        columns = row.find_all('td')
        day_period = columns[0].text.strip()  # Sáng, Chiều, Tối
        
        for i, column in enumerate(columns[1:], start=2):  # Bắt đầu từ Thứ 2 đến Chủ nhật
            day_of_week = f"Thứ {i}" if i < 8 else "Chủ nhật"
            classes = column.find_all('div', class_='content')
            
            for class_ in classes:
                class_info = {
                    'Tên môn học': class_.find('a').text.strip(),
                    'Mã lớp': class_.find_all('p')[0].text.strip(),
                    'Tiết': class_.find_all('p')[1].text.strip().split('\n')[0].replace('Tiết:', '').strip(),
                    'Phòng': class_.find_all('p')[1].find('span').text.strip(),
                    'Giảng viên': class_.find_all('p')[2].text.strip().replace('GV:', '').strip()
                }
                if day_of_week not in schedule:
                    schedule[day_of_week] = {day_period: [class_info]}
                else:
                    if day_period in schedule[day_of_week]:
                        schedule[day_of_week][day_period].append(class_info)
                    else:
                        schedule[day_of_week][day_period] = [class_info]
    return schedule

class Post(scrapy.Item):
    pass

class MySpider(scrapy.Spider):
    name = 'my_spider'
    def start_requests(self):
        url = "https://sv.iuh.edu.vn/SinhVienTraCuu/GetDanhSachLichTheoTuan"
        formdata = {
            "k": "k=nzmlGyKmW2SCp8gVv0i-wxVEX2HvAGhYxRxBx38PXog"
        }
        yield scrapy.http.FormRequest(url=url, callback=self.parse, formdata=formdata)

    def parse(self, response):
        # create a html of url
        soup = BeautifulSoup(response.text, 'html.parser')
        # get schedule
        schedule = soup_to_scheduler(soup)
        with open('schedule_3.json', 'w', encoding='utf-8') as f:
            json.dump(schedule, f, ensure_ascii=False, indent=4)
        

```

### Bài 4: 
Sinh viên dùng 01 thư viện trong các thư viện: urllib và BeautifulSoup, Selenium hay Scrapy để tiến hành cào dữ liệu với các yêu cầu sau:
<br> 4.1. Sử dụng đường link: https://cafeland.vn/du-an/
<br> 4.2. Thực hiện các tác vụ lấy thông tin các bài viết về dự án bất động sản.
<br> Các thông tin bao gồm :
* ✓ Loại dự án
* ✓ Tên dự án
* ✓ Vị trí tọa độ của công trình trên map
* ✓ Thông tin nhanh của dự á

```python
import requests
from bs4 import BeautifulSoup
import datetime
import pandas as pd
import csv
import time

headers={'User-Agent': 'Bot1'}


def list_link():
    return [f"https://cafeland.vn/du-an/page-{item}/" for item in range(1,195)]

def get_link(url):
    r = requests.get(url, headers=headers)
    if r.status_code == 200:
        return BeautifulSoup(r.text, 'lxml')
    else:
        print("Failed to fetch page: ", r.status_code)
        return None

def get_page_link(list_link):
    list_link_products = []
    for url in list_link[:3]:
        time.sleep(0.1)
        soup = get_link(url)
        if soup:
            list_link_for_page = soup.find('ul', class_='list-type-14 pt15 listChild imgCustLoad projectList')
            if list_link_for_page:
                for s in list_link_for_page.select("h3 > a[href^='https://cafeland.vn/du-an']"):
                    list_link_products.append(s['href'])
            else:
                print("List link for page not found")
        else:
            print("Failed to fetch page")
    return list_link_products

def access_link_product(link):
    soup = get_link(link)
    time.sleep(0.2)
    if soup:
        table = soup.tbody
        rows = table.select("tr")
        res = {}
        for r in rows:
            cells = r.select("td")
            for c in cells:
                key = c.strong.text.replace(":", "").strip()
                span = c.text.replace(key, "").replace(":", "").strip()
                res[key] = span
                
        return res
    return None
    
def auto_access_link(links):
    products = []
    for item in links:
        print(item)
        try:
            product = access_link_product(item)
            if not product:
                continue 
            products.append(product)
        except Exception as e:
            print(e)
            print("Error with link: ", item)
    return products

def output(products):
    # Chuyển list sang DataFrame
    data = pd.DataFrame.from_records(products)
    data.to_csv('data.csv', index=False, encoding='utf-8-sig')
    print("Data saved")

def main():
    list_links = list_link()
    links = get_page_link(list_links)
    products = auto_access_link(links)
    output(products)

if __name__ == "__main__":
    main()
    
```
