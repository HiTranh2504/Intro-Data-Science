```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time

service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

url = "https://sv.iuh.edu.vn/tra-cuu/lich-hoc-theo-tuan.html?k=7-tVfq_Qc2TzHK9PGJKyfDjZnsDmKGEy0dBmdrBjifM"

driver.get(url)

time.sleep(5)  

html_source = driver.page_source

driver.quit() # Đóng trình duyệt

soup = BeautifulSoup(html_source, 'html.parser')

# Cấu trúc để lưu lịch
schedule = {}

for row in soup.find_all('tr')[1:]:  
    columns = row.find_all('td')
    day_period = columns[0].text.strip()  
    
    for i, column in enumerate(columns[1:], start=2): 
        day_of_week = f"Thứ {i}" if i < 8 else "Chủ nhật"
        classes = column.find_all('div', class_='content')
        
        for class_ in classes:
            class_info = {
                'Tên môn học': class_.find('a').text.strip(),
                'Mã lớp': class_.find_all('p')[0].text.strip(),
                'Tiết': class_.find_all('p')[1].text.strip().split('\n')[0].replace('Tiết:', '').strip(),
                'Phòng': class_.find_all('p')[1].find('span').text.strip(),
                'Giảng viên': class_.find_all('p')[2].text.strip().replace('GV:', '').strip()
            }
            if day_of_week not in schedule:
                schedule[day_of_week] = {day_period: [class_info]}
            else:
                if day_period in schedule[day_of_week]:
                    schedule[day_of_week][day_period].append(class_info)
                else:
                    schedule[day_of_week][day_period] = [class_info]


import pandas as pd

# Chuẩn bị dữ liệu
data = []
for day, info in schedule.items():
    print(info)
    for period, classes in info.items():
        
        for class_info in classes:
            class_info['Thời gian'] = period
            class_info['Ngày'] = day
            class_info['Tiết'] = class_info['Tiết'].replace('-', 'to')

            data.append(class_info)

df = pd.DataFrame(data).astype(str)

#Lưu dataframe sang file csv
df.to_csv('lich_hoc2.csv', index=False, encoding='utf-8-sig')

print("Thành công")
print("_____________________________________________________________________________________________________________________")

print(df.head())
```

```python
import scrapy
from scrapy.http import FormRequest
from scrapy.selector import Selector
from PIL import Image
import pytesseract

class MySpider(scrapy.Spider):
    name = 'myspider'
    start_urls = ['https://sv.iuh.edu.vn/sinh-vien-dang-nhap.html']

    def parse(self, response):
        # Nhận diện và giải quyết CAPTCHA
        captcha_url = response.css('#imgCaptcha::attr(src)').get()
        captcha_text = self.solve_captcha(captcha_url)

        # Đăng nhập
        yield FormRequest.from_response(
            response,
            formid='login_form',
            formdata={
                'txtTaiKhoan': '22655141',
                'txtMatKhau': '2011',
                'txtCaptcha': captcha_text
            },
            callback=self.after_login
        )

    def after_login(self, response):
        if "Đăng nhập thành công" in response.text:
            self.logger.info("Đăng nhập thành công!")
            # Tiếp tục crawl thời khóa biểu ở đây
        else:
            self.logger.error("Đăng nhập không thành công!")

    def solve_captcha(self, captcha_url):
        # Tải ảnh CAPTCHA từ URL và nhận diện văn bản
        img = Image.open(captcha_url)
        captcha_text = pytesseract.image_to_string(img)
        return captcha_text
```
