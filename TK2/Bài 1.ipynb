### Bài 1:
Cho website phim imdb ở đường link: https://www.imdb.com/ Hãy thực hiện yêu cầu cào ít nhất 20 bộ phim và lưu vào 1 file csv với các cột tương ứng với các thông tin: Số thứ tự(sinh tự động), Tên bộ phim, Đạo diễn, Biên kịch, Diễn viên, Giới thiệu bộ phim. (Nếu có nhiều người thì tên mỗi người cách nhau bởi dấu ,).

```python
from bs4 import BeautifulSoup
import requests

headers = {'User-Agent':'Bot1'}

url ='https://www.imdb.com/chart/top/?ref_=nv_mv_250'
response = requests.get(url, headers=headers)
html = response.text
soup = BeautifulSoup(html, 'html.parser')

#Lấy link từng bộ phim
Links = soup.find('ul', class_='ipc-metadata-list ipc-metadata-list--dividers-between sc-a1e81754-0 eBRbsI compact-list-view ipc-metadata-list--base')
#print(Links)
LinkArr = []
count = 0
for item in Links.find_all('li'):
    head = 'https://www.imdb.com/'
    link = head + item.find('a')['href']
    LinkArr.append(link)
    count+=1
    if count == 20:
        break
    
print(LinkArr)
```

```python
Film_Arr = []
for i in LinkArr:
    headers = {'User-Agent':'Bot1'}
    url = i
    response = requests.get(url, headers=headers)
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')

    #Lấy nội dung của từng bộ phim
    Arr = [] # Mảng chứa thông tin của bộ phim (tp,dd, bk, dv, gt)
    #Ten phim
    Name = soup.find('div',class_='sc-4e4cc5f9-3 dDRspk').find('h1').find('span',class_='hero__primary-text').text
    Arr.append(Name)
    

    link = soup.find('div', class_='sc-b7c53eda-2 iOESUA').find('div', role='presentation').find('ul', class_='ipc-metadata-list ipc-metadata-list--dividers-all title-pc-list ipc-metadata-list--baseAlt').find('li', class_='ipc-metadata-list__item ipc-metadata-list-item--link').find('a')['href']
    lienket = 'https://www.imdb.com/' + link


    def get_detail(lienket):
        #Lấy thông tin đạo diễn, diễn viên, biên kịch của bộ phim
        header = {'User-Agent':'Bot1'}
        url = lienket
        response = requests.get(url, headers=header)
        soup = BeautifulSoup(response.text, 'html.parser')


        #Đạo diễn
        Director = soup.find('div',id='fullcredits_content').find('table',class_='simpleTable simpleCreditsTable').find_all('tr')
        Daodien = []
        for item in Director:
            daodien = item.find('td',class_='name').text.strip()
            Daodien.append(daodien)
        Arr.append(Daodien)
        # Biên kịch 
        # table 28
        h4_tag = soup.find('h4', {'name': 'script_department', 'class': 'dataHeaderWithBorder'})
        table_tag = h4_tag.find_next('table', class_='simpleTable simpleCreditsTable')
        BKich = table_tag.find_all('tr')
        Bienkich = []
        for item in BKich:
            bk = item.find('td',class_='name').text.strip()
            Bienkich.append(bk)
        Arr.append(Bienkich)

        #Diễn viên
        Actor = soup.find('div', id='fullcredits_content').find('table', class_='cast_list').find_all('tr')
        Dienvien = []
        for item in Actor[1:]:
            ac = item.find_all('td')
            if len(ac) > 1:
                a = ac[1].find('a').text.strip()
                Dienvien.append(a)
        Arr.append(Dienvien)


    get_detail(lienket)

    #Giới thiệu phim
    Intro = soup.find('div',class_='sc-4e4cc5f9-4 ixQHwZ').find('div',class_='sc-4e4cc5f9-6 fFWQTt').find('span',role='presentation').text.strip()
    Arr.append(Intro)   
    
    #Chuyển mảng thành dictionary
    Film_dict = {
        'Tên phim': (Arr[0]),
        'Đạo diễn': ','.join(Arr[1]),
        'Biên kịch':','.join(Arr[2]),
        'Diễn viên':','.join(Arr[3]),
        'Giới thiệu film':','.join(Arr[4])
        
    }
    Film_Arr.append(Film_dict)


print(Film_Arr)
```

```python
import pandas as pd
import csv

df = pd.DataFrame(Film_Arr)
df.to_csv('film_data.csv', index=True, encoding='utf-8-sig')
```
